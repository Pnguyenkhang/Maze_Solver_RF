```python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Load in maze from file
def load_maze(mazename):
    return np.load(mazename)
maze = load_maze("maze4by4.npy") # testing

# x = j
# y = i 
# Boundary Setting
L = 3 # length is also x,y of matrix that we will be using
win_x, win_y = 2,2
death_x, death_y = 0,2

# Constants defining the path moves
RIGHT = 0
LEFT = 1
DOWN = 2
UP = 3

# Q_table creation
q_table = np.zeros((L,L,4), int)
env_matrix = np.zeros((L,L,4), int)

# fills the edges of the q-table with punishment values
def fill_q_table(L, q_table):
    for y in range(L):
        for x in range(L):
            if x == 0: # up // left on 2d array
                q_table[y,x,LEFT] = -1000
                #print(q_table[y,x])
            if y == 0: # right // up on 2d
                q_table[y,x,UP] = -1000
                #print(q_table[y,x])
            if x == L-1: # left // right
                q_table[y,x,RIGHT] = -1000
                #print(q_table[y,x])
            if y == L-1: # right // down
                q_table[y,x,DOWN] = -1000
                #print(q_table[y,x])


            # for obstacle placement, we treat them as badly as wall boundaries
            if maze[y,x] == -100:
                # boundary checker
                if y > 1: # down
                    q_table[y-1,x,DOWN] = -100

                if y < L-1:# up
                    q_table[y+1,x,UP] = -100

                if x > 1: # right
                    q_table[y,x-1,RIGHT] = -100

                if x < L-1: #left
                    q_table[y,x+1,LEFT] = -100


    # for winning spot
    q_table[win_x-1,win_y,DOWN] = 100
    #q_table[win_x+1,win_y,UP] = 100
    q_table[win_x,win_y-1,RIGHT] = 100
    #q_table[win_x,win_y+1,LEFT] = 100
    return q_table
```

```python
def print_q_table(q_table):
    count = 0
    print("           y  x   right left down up")
    for y in range(L):
        for x in range(L):
            count+=1
            print(count, 'coords:',(y,x),q_table[y,x])
            
```

```python
def getAllPossibleMove(cur_x, cur_y):
    # list comprehension check on positioning
    poss_step = [x != -1000 for x in q_table[cur_x,cur_y]]
    action = []
    #print(poss_step)
    if(poss_step[0]):
        action.append(RIGHT)
    if(poss_step[1]):
        action.append(LEFT)
    if(poss_step[2]):
        action.append(DOWN)
    if(poss_step[3]):
        action.append(UP)
    #print('a',action)
    return action


def getWinningNextAction(cur_x, cur_y, q_table):
    max_val = q_table[cur_x,cur_y]
    print(max_val)
    max_index = np.argmax(max_val)
    print(max_index)
    return max_index


# x is vertical downwards orientation
# y is horizontal rightwards orientation

def getNextState(cur_x,cur_y, action):
    if(action == LEFT):
        print("LEFT")
        return cur_x,cur_y-1
    if(action == RIGHT):
        print("RIGHT")
        return cur_x,cur_y+1
    if(action == UP):
        print("UP")
        return cur_x-1,cur_y
    if(action == DOWN):
        print("DOWN")
        return cur_x+1,cur_y
    
def CheckGameOver(cur_x, cur_y, maze):
    print(maze)
    if cur_x == win_x and cur_y == win_y:
        print("WIN!")
        return True

    elif cur_x == death_x and cur_y == death_y:
        print("DEATH!")
        return True

    else:
        print("no gameover yet!")
        return False
# Test movement
cur_x = 0
cur_y = 0


def bound_checker(cur_x,cur_y, action):
    if cur_x == 0 and action == UP:
        return True
    if cur_x == L-1 and action == DOWN:
        return True
    if cur_y == 0 and action == LEFT:
        return True
    if cur_y == L-1 and action == RIGHT:
        return True
    else:
        return False

getWinningNextAction(cur_x, cur_y, q_table)
```

```python
# fills q_table with the dangers
env_matrix = fill_q_table(L,env_matrix)
print_q_table(env_matrix)
```

```python
# starting loop
discount = 0.9
learning_rate = 0.1
epsilon = 0.7 # ratio at which the agent will explore vs exploit

# animation
maze_frames = list()


for episode in range(100):
    print("----------Start of Episode:",episode,'----------')
    
    # refreshes the maze each time
    visual_maze = load_maze("maze3by3.npy")
    
    # random beginning point generation
    cur_x = random.randint(0,L-1)
    cur_y = random.randint(0,L-1)
    
    # End Goal
    visual_maze[win_x,win_y] = 100
    visual_maze[cur_x,cur_y] = 1000
    #print(visual_maze)
    # main trial error loop
    while(not CheckGameOver(cur_x, cur_y, visual_maze)):
        print("Ingame")
        if episode % 20 == 0:
            maze_frames.append(visual_maze)
        plt.imshow(visual_maze, vmin = -100, vmax = 1000)
        plt.show()
        # pick from random choice
        choices = getAllPossibleMove(cur_x, cur_y)
        choice = random.choice(choices)
        while(bound_checker(cur_x,cur_y, choice)):
            choice = random.choice(choices)
            
        new_x, new_y = getNextState(cur_x,cur_y, choice)
        
        
        # Bellman equation
        lr1 = q_table[cur_x,cur_y, choice]
        lr2 = learning_rate * (env_matrix[cur_x,cur_y,choice] + discount*max(q_table[new_x,new_y]) - q_table[cur_x,cur_y,choice])
    
        q_table[cur_x,cur_y,choice] = lr1 + lr2
        
        # visual movement of the agent
        visual_maze[cur_x, cur_y] = 0
        cur_x = new_x
        cur_y = new_y
        visual_maze[cur_x, cur_y] = 1000
        
    print('----------Final Move----------')
    plt.imshow(visual_maze, vmin = -100, vmax = 1000)
    plt.show()
    if episode % 20 == 0:
            maze_frames.append(visual_maze)
```

```python
# Using Trained solution
for episode in range(10):
    print("----------Start of Episode:",episode,'----------')
    
    # refreshes the maze each time
    visual_maze = load_maze("maze3by3.npy")
    
    # random beginning point generation
    cur_x = random.randint(0,L-1)
    cur_y = random.randint(0,L-1)
    
    # End Goal
    visual_maze[2,2] = 100
    visual_maze[cur_x,cur_y] = 1000
    #print(visual_maze)
    # main trial error loop
    while(not CheckGameOver(cur_x, cur_y, visual_maze)):
        print("Ingame")
        plt.imshow(visual_maze, vmin = -100, vmax = 1000)
        plt.show()
        
        choice = getWinningNextAction(cur_x, cur_y, q_table)
            
        new_x, new_y = getNextState(cur_x,cur_y, choice)
        
        
        # Bellman equation
        lr1 = q_table[cur_x,cur_y, choice]
        lr2 = learning_rate * (env_matrix[cur_x,cur_y,choice] + discount*max(q_table[new_x,new_y]) - q_table[cur_x,cur_y,choice])
    
        q_table[cur_x,cur_y,choice] = lr1 + lr2
        
        # visual movement of the agent
        visual_maze[cur_x, cur_y] = 0
        cur_x = new_x
        cur_y = new_y
        visual_maze[cur_x, cur_y] = 1000
        
    print('----------Final Move----------')
    plt.imshow(visual_maze, vmin = -100, vmax = 1000)
    plt.show()
```
